## Q4.G

Null Hypothesis H0: Proportion of companies operating between winners and contestants are equal

Alternate Hypothesis H1 : Proportion of companies operating between winners and contestants different. 

```python
Total_winner = D_Doperating[D['Result'] == 'Winner'].Result.value_counts()[0]  
Total_Contestants = D_Doperating[D_Doperating['Result'] == 'Contestant'].Result.value_counts()[0] 

Total_DWinners = D_DNEW[D_DNEW['Result'] == 'Winner'].Result.value_counts()[0]
Total_DContestants = D_DNEW[D_DNEW['Result'] == 'Contestant'].Result.value_counts()[0]

print ('Total winners who are operating are =', Total_winner)
print ('Total contestants who are operating are =', Total_Contestants)
print ('Total winners are =', Total_DWinners)
print ('Total contestants are =', Total_DContestants)
```

## Q4.H

```python
t_value,p_value=stats.ttest_ind([Total_winner, Total_Contestants],[Total_DWinners, Total_DContestants])

print('Test statistic is %f'%float("{:.6f}".format(t_value)))

print('p-value for two tailed test is %f'%p_value)

alpha = 0.05

if p_value<=alpha:

    print(f'With a p-value of {round(p_value,4)} the difference is significant. Null hypothesis is rejected')
else:

    print(f'With a p-value of {round(p_value,4)} the difference is not significant, hence We fail to reject null hypothesis')
```
